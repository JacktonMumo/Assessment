```python
import pandas
```

```python
customer = pandas.read_csv('customerdata.csv')
customer
```

```python
customer.isnull().sum()
```

```python
array=customer.values
array
```

```python
x=array[0:661,2:7]
x
```

```python
from sklearn.cluster import KMeans as km
model=km(n_clusters=12,random_state=42)
model.fit(x)
```

```python
groups=model.cluster_centers_
groups
```

```python
clusters=pandas.DataFrame(groups,columns=['Avg_Credit_Limit','Total_Credit_Cards','Total_visits_bank','Total_visits_online','Total_calls_made'])
clusters
```

```python
customer['group_number']=model.labels_
customer.head(10)
```

```python
customer.groupby('group_number').size()
```
